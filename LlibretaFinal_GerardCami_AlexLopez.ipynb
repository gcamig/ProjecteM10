```python
# Importem les llibreries necessàries
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importem els arxius .csv amb els datasets que utilitzarem
equipos = pd.read_csv('stats_equipos.csv', encoding='latin-1')
jugadores = pd.read_csv('stats_jugadores.csv', encoding='latin-1')
```

```python
# Definim el dataframe 'stats_equipos' i 'stats_jugadores'
stats_equipos = pd.DataFrame(equipos)
stats_jugadores = pd.DataFrame(jugadores)
```

```python
# Mostrem les columnes que té el dataframe que hem definit prèviament 'stats_equipos'
stats_equipos.columns
```

```python
# Esborrem les columnes innecessàries
del stats_equipos['Unnamed: 0']
```

```python
# Modifiquem el nom de totes les columnes
stats_equipos = stats_equipos.rename(columns={"Teams":"Equipos"})
stats_equipos = stats_equipos.rename(columns={"MP":"Partidos Jugados"})
stats_equipos = stats_equipos.rename(columns={"W":"Victorias"})
stats_equipos = stats_equipos.rename(columns={"D":"Empates"})
stats_equipos = stats_equipos.rename(columns={"L":"Derrotas"})
stats_equipos = stats_equipos.rename(columns={"GF":"Goles a favor"})
stats_equipos = stats_equipos.rename(columns={"GA":"Goles en contra"})
stats_equipos = stats_equipos.rename(columns={"+/-":"Diferencia de goles"})
stats_equipos = stats_equipos.rename(columns={"Pts":"Puntos"})
stats_equipos = stats_equipos.rename(columns={"Group":"Grupo"})
```

```python
# Mostrem com ha quedat el dataframe
stats_equipos.head(32)
```

```python
# Mostrem les columnes que té el dataframe que hem definit prèviament 'stats_jugadores'
stats_jugadores.columns
```

```python
#Esborrem camps innecessàris
del stats_jugadores['FIFA Ranking ']
del stats_jugadores['National Team Kit Sponsor']
del stats_jugadores['National Team Jersey Number']
del stats_jugadores['Player DOB']
del stats_jugadores['Club ']
del stats_jugadores[' Appearances']
del stats_jugadores['Assists Provided ']
del stats_jugadores['Dribbles per 90']
del stats_jugadores['Interceptions per 90']
del stats_jugadores['Tackles per 90']
del stats_jugadores['Total Duels Won per 90']
del stats_jugadores['Save Percentage']
del stats_jugadores['Clean Sheets']
del stats_jugadores['Brand Sponsor/Brand Used']
```

```python
#Actualitzem els nom dels camps útils per utilitzar-los amb facilitat
stats_jugadores = stats_jugadores.rename(columns={'Nationality ':'Equipos'})
stats_jugadores = stats_jugadores.rename(columns={'Position ':'Posiciones'})
stats_jugadores = stats_jugadores.rename(columns={'Player Name ':'Jugadores'})
stats_jugadores = stats_jugadores.rename(columns={'Goals Scored ':'GolesMarcados'})
```

```python
# Mostrem com ha quedat el dataframe
jugadores1 = stats_jugadores.head(50)
jugadores1.head(50)
```

```python
#Cambiar el tipus de dada del camp GolesMarcados
#Trec totes les entrades coincidents amb - substituint el caracter per un 0
jugadores1['GolesMarcados'] = jugadores1['GolesMarcados'].str.replace('-','0')
#Faig una conversio a int de tots els elements de la columna
jugadores1 = jugadores1.astype({'GolesMarcados':'int'})
```

```python
for x in jugadores1['GolesMarcados']:
    print(x)
    print(type(x))
```

```python
jugadores_ordenados = jugadores1.sort_values(by='GolesMarcados', ascending=False)
```

```python
jugadores_ordenados.head(50)
```

```python
#Definim un dataframe que guardi els 4 jugadors amb més gols
jugadores2 = jugadores_ordenados.head(4)
```

```python
#Mostrem els jugadors en un gràfic de barres horitzontals
jugadores2.plot(kind='barh', y='GolesMarcados', x='Jugadores')
```
